### Raw text to Tokens
- Input ID
- Tokens

```python
from transformers import AutoTokenizer

#used to refer to a specific fine-tuned version of the DistilBERT model for sentiment analysis on the Stanford Sentiment Treebank (SST-2) dataset in English.
checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint) #method
```

```python
raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt") #To specify the type of tensors-return_tensors , To set the length-paddding

print("...PT INPUTS ...\n")
print(inputs)
print("\n... TOKENIZER ...\n")
print(tokenizer)
print("\n... RAW INPUTS ...")
print(raw_inputs)
```

### Input Ids to Logits
- output shape
- Automodel
- AutoModelForSequenceClassification

```python
from transformers import AutoModel

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModel.from_pretrained(checkpoint)
outputs = model(**inputs)
print("TF OUTPUT SHAPE ...\n")
print(outputs.last_hidden_state.shape)
#Tensor has 2 sentences each has 16 tokens and hidden size of the model is 768
```

- for classification problem

```python
from transformers import AutoModelForSequenceClassification

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
model = AutoModelForSequenceClassification.from_pretrained(checkpoint)
outputs = model(**inputs)

print("TF OUTPUT SHAPE ...\n")
print(outputs.logits.shape)
```

```python
#Logits , the raw, unnormalized scores outputted by the last layer of the model . Sum of these vlues is not 1
print(outputs.logits)
```

### Logits to Prediction
- Use SOFTMAX to get the probability from logits
- To get the labels corresponding to each position

```python
import torch

predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)# Calculate the softmax of the output logits

print("PT OUTPUT PROBABILITIES ...\n")
print(predictions)

print("\n\n... PT OUTPUT PROBABILITIES WITH LABEL TITLES ...\n")
for i, pred in enumerate(predictions): print(f"Sentence #{i+1} - Classification Probabilities:\n\t{model.config.id2label[0]}: %{100*pred[0]:.3f}\n\t{model.config.id2label[1]}: %{100*pred[1]:.3f}")

print("\n\n... MODEL LABEL ...\n")
print(model.config.id2label)
```

# Putting it all together

```python
from transformers import AutoTokenizer

checkpoint = "distilbert-base-uncased-finetuned-sst-2-english"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)

sequence = "I've been waiting for a HuggingFace course my whole life."
sequences_same_length = ["Been holding out for a HuggingFace course my entire life.", "I've waited for a HuggingFace course so long!"]
sequences_diff_length = ["I've been waiting for a HuggingFace course my whole life.", "So have I!"]

print("... WHAT CAN THE TOKENIZER OBJECT DO? ...\n")
print ("TOKENIZE A SINGLE SEQUENCE")
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequence, return_tensors='tf')}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequence, return_tensors='pt')}")

print("\n\nTOKENIZE MULTIPLE SEQUENCES - SAME LENGTH")
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_same_length, return_tensors='tf')}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_same_length, return_tensors='pt')}")

print("\n\nTOKENIZE MULTIPLE SEQUENCES - DIFFERENT LENGTH")
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='tf', padding=True)}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='pt', padding=True)}")

print("\n\nPADDING UP TO SEQUENCE MAX LENGTH") # Will pad the sequences up to the maximum sequence length
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='tf', padding='longest')}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='pt', padding='longest')}")

print("\n\nPADDING UP TO MODEL MAX LENGTH")
# Will pad the sequences up to the model max length (512 for BERT or DistilBERT)
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='tf', padding='max_length')}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='pt', padding='max_length')}")

print("\n\nPADDING UP TO SPECIFIED LENGTH")# Will pad the sequences up to the specified max length
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='tf', padding='max_length', max_length=8, truncation=True)}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='pt', padding='max_length', max_length=8, truncation=True)}")

print("\n\nTRUNCATE TO MODEL MAX LENGTH")
# Will truncate the sequences that are longer than the model max length(512 for BERT or DistilBERT)
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='tf', truncation=True, padding=True)}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='pt', truncation=True, padding=True)}")

print("\n\nTRUNCATE TO A SPECIFIED LENGTH")# Will truncate the sequences that are longer than the specified max length
print("(RETURN TF COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='tf', padding=True, truncation=True, max_length=8)}")
print("(RETURN PT COMPATIBLE TENSORS)")
print(f"\t--> {tokenizer(sequences_diff_length, return_tensors='pt', padding=True, truncation=True, max_length=8)}")
```

## Preprocessing Sentence Pairs

```python
from transformers import AutoTokenizer

checkpoint = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(checkpoint)
tokenized_sentences_1 = tokenizer(raw_datasets["train"]["sentence1"])
tokenized_sentences_2 = tokenizer(raw_datasets["train"]["sentence2"])

print("FIRST 3 ORIGINAL SENTENCE PAIRS")
for i in range(3): print(f"\tSENTENCE 1 --> '{raw_datasets['train']['sentence1'][i]}'\n\tSENTENCE 2 --> '{raw_datasets['train']['sentence2'][i]}'\n")

print("\nFIRST 3 TOKENIZED SENTENCE PAIRS")
for i in range(3): print(f"\tSENTENCE 1 --> '{tokenized_sentences_1[i]}'\n\tSENTENCE 2 --> '{tokenized_sentences_2[i]}'\n")

e15_s1 = raw_datasets['train']['sentence1'][15]
e15_s2 = raw_datasets['train']['sentence2'][15]

print("\nELEMENT 15 - ORIGINAL SENTENCE PAIRS")
print(f"\tSENTENCE 1 --> '{e15_s1}'\n\tSENTENCE 2 --> '{e15_s2}'\n")

e15_tokenize_seperately_s1 = tokenizer.convert_ids_to_tokens(tokenizer(e15_s1)["input_ids"])
e15_tokenize_seperately_s2 = tokenizer.convert_ids_to_tokens(tokenizer(e15_s2)["input_ids"])
e15_tokenized_together = tokenizer(e15_s1, e15_s2)
e15_tokens = tokenizer.convert_ids_to_tokens(e15_tokenized_together["input_ids"])

print("\nELEMENT 15 - TOKENIZED SEPERATELY")
print(f"\tSENTENCE 1 --> {' '.join(e15_tokenize_seperately_s1)}\n\tSENTENCE 2 --> {' '.join(e15_tokenize_seperately_s2)}\n")

print("\nELEMENT 15 - TOKENIZED TOGETHER")
print(f"\tSENTENCES --> {' '.join(e15_tokens)}\n")
```
